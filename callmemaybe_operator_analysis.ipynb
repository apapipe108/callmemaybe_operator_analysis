# Carga de datos.
Comenzamos cargando los datos e imprimiendo las primeras filas para dar un primer vistazo. 

```python
# Importamos las librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Cargamos los datasets
telecom_data= pd.read_csv('telecom_dataset_us.csv')
clients_data= pd.read_csv('telecom_clients_us.csv')

# Mostramos las primeras filas de cada dataset
print("Dataset de llamadas:")
display(telecom_data.head())

print("\nDataset de clientes:")
display(clients_data.head())

# Información general sobre los datasets
print("\nInfo general de llamadas:")
telecom_data.info()

print("\nInfo general de clientes:")
clients_data.info()

```

# Limpieza de datos 

A continuacion revisaremos que los datos se encuentren en orden y que puedan dar fidelidad a nuestros analisis. 


# Valores nulos

```python
# Revisamos valores nulos
print("\nValores nulos en llamadas:")
print(telecom_data.isnull().sum())

print("\nValores nulos en clientes:")
print(clients_data.isnull().sum())
```

```python
# Verificar valores nulos en calls_df
print("Valores nulos en telecom_data:")
print(telecom_data.isnull().sum())

# Verificar valores nulos en clients_df
print("\nValores nulos en clients_data:")
print(clients_data.isnull().sum())

# Contar la cantidad de llamadas internas y externas
internal_counts = telecom_data['internal'].value_counts()

# Mostrar la proporción
print("Proporción de llamadas internas y externas:")
print(internal_counts)

# Reemplazar valores nulos en 'internal' con False (asumiendo llamadas externas)
telecom_data['internal'] = telecom_data['internal'].fillna(False)

# Reemplazar valores nulos en 'operator_id' con 'UNKNOWN'
telecom_data['operator_id'] = telecom_data['operator_id'].fillna('UNKNOWN')

# Verificar que ya no haya valores nulos
print("\nValores nulos después de limpieza:")
print(telecom_data.isnull().sum())
```

Hemos encontrado valores nulos en "telecom data" en las columnas siguientes: "internal" (117 valores faltantes) y en "operator_id" (más de 8,000 valores faltantes). Mientras que en "clients_data" no encontramos ningun valor nulo.

Los valores nulos en internal (117) son una cantidad muy pequeña en comparación con el total de llamadas (53,785). Esto significa que reemplazarlos no tendrá un impacto significativo en el análisis. Las llamadas externas representan casi el 89% del total de los datos, por lo que convertir estos pocos valores nulos a este grupo es una buena manera de manejar estos datos.

Los valores nulos en opertaror id son 8172, esta columna es crucial para identificar a los operadores y analizar su desempeño (por ejemplo, llamadas perdidas, tiempo de espera, etc.). Estos valores nulos no necesariamente significan que los datos sean inválidos. Podrían representar llamadas que no están asociadas a un operador específico. Dado que eliminar 8,172 filas podría afectar negativamente el análisis, optaremos por reemplazar los valores nulos con "UNKNOWN". Esto nos permitirá mantener todas las filas en el dataset, asi como identificar las llamadas no asociadas a un operador específico y asi poder realizar nuestro analisis sin perder información.

# Valores duplicados 

```python
# Verificar duplicados en telecom_data
print("Filas duplicadas en telecom_data:", telecom_data.duplicated().sum())

# Verificar duplicados en clients_data
print("Filas duplicadas en clients_data:", clients_data.duplicated().sum())
```

```python
# Identificar duplicados exactos
duplicados_exactos = telecom_data[telecom_data.duplicated(keep=False)]  # keep=False marca todos los duplicados

print("Ejemplo de filas duplicadas exactas:")
print(duplicados_exactos.head(10))
```

```python
# Eliminar duplicados exactos
telecom_data = telecom_data.drop_duplicates()

# Verificar el número de filas después de eliminar duplicados
print("Número de filas después de eliminar duplicados:", len(telecom_data))
```

Hemos encontrado que hay 4900 filas duplicadas en el "telecom data" que represental 9.1% del total de datos.
Para decidir como tratar con estos datos duplicados he realizado un analisis de algunas de las filas duplicadas, para poder determinar si estos duplicados son fallos del sistema o si son llamadas validas que simplemente se realizaron varias veces. 
Todas las filas duplicadas que observe son idénticas en todas las columnas, lo que sugiere que son errores (por ejemplo, llamadas registradas dos veces por un fallo técnico)
No hay diferencias sutiles que indiquen que sean llamadas válidas que ocurrieron más de una vez.
Considerando la informacion antes mencionada determine que la mejor opción es eliminarlos. Esto garantizará que el análisis sea preciso y no esté distorsionado por datos repetidos.


# Llamadas con duracion 0 

```python
# Filtrar llamadas con duración 0
llamadas_0_duration = telecom_data[
    (telecom_data['call_duration'] == 0) & 
    (telecom_data['total_call_duration'] == 0)
]

# Verificar cuántas están marcadas como llamadas perdidas
num_total_0 = len(llamadas_0_duration)
num_missed_0 = llamadas_0_duration['is_missed_call'].sum()

# Mostrar resultados
print(f"Número de llamadas con duración 0: {num_total_0}")
print(f"Número de llamadas con duración 0 y marcadas como perdidas: {num_missed_0}")
print(f"Porcentaje de llamadas con duración 0 que fueron perdidas: {num_missed_0 / num_total_0 * 100:.2f}%")
```

Hemos identificado 1195 llamadas con duración igual a cero, tanto en el tiempo real de conversación como en el total (incluyendo tiempo de espera). Tras un análisis más detallado, observamos que el 99.8% de estas llamadas (1193) están marcadas como llamadas perdidas. Esto confirma que no se trata de errores de registro, sino de intentos de llamada fallidos. Por lo tanto, decidimos mantenerlas en el análisis, ya que aportan información valiosa sobre la eficacia de los operadores.

# Valores atipicos

```python
# Estadísticas descriptivas de columnas numéricas
print("Estadísticas de calls_count:")
print(telecom_data['calls_count'].describe())

print("\nEstadísticas de call_duration:")
print(telecom_data['call_duration'].describe())

print("\nEstadísticas de total_call_duration:")
print(telecom_data['total_call_duration'].describe())
```

```python
# Calcular el IQR para cada columna numérica
Q1 = telecom_data[['calls_count', 'call_duration', 'total_call_duration']].quantile(0.25)
Q3 = telecom_data[['calls_count', 'call_duration', 'total_call_duration']].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites inferior y superior
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

print("Límites inferiores:")
print(limite_inferior)

print("\nLímites superiores:")
print(limite_superior)
```

```python
# Identificar valores atípicos en cada columna
outliers_calls_count = telecom_data[
    (telecom_data['calls_count'] < limite_inferior['calls_count']) | 
    (telecom_data['calls_count'] > limite_superior['calls_count'])
]

outliers_call_duration = telecom_data[
    (telecom_data['call_duration'] < limite_inferior['call_duration']) | 
    (telecom_data['call_duration'] > limite_superior['call_duration'])
]

outliers_total_call_duration = telecom_data[
    (telecom_data['total_call_duration'] < limite_inferior['total_call_duration']) | 
    (telecom_data['total_call_duration'] > limite_superior['total_call_duration'])
]

# Mostrar la cantidad de valores atípicos
print("Valores atípicos en calls_count:", len(outliers_calls_count))
print("Valores atípicos en call_duration:", len(outliers_call_duration))
print("Valores atípicos en total_call_duration:", len(outliers_total_call_duration))
```

```python
# Gráficos de caja para las columnas numéricas
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.boxplot(y=telecom_data['calls_count'])
plt.title('Boxplot de calls_count')

plt.subplot(1, 3, 2)
sns.boxplot(y=telecom_data['call_duration'])
plt.title('Boxplot de call_duration')

plt.subplot(1, 3, 3)
sns.boxplot(y=telecom_data['total_call_duration'])
plt.title('Boxplot de total_call_duration')

plt.show()
```

El mayor reto en la limpieza de datos vino al descubrir los valores atipicos, que eran: Valores atípicos en calls_count: 6047,
Valores atípicos en call_duration: 6866, Valores atípicos en total_call_duration: 5932.

Debo de realizar un analisis mas profundo para conocer la naturaleza de estos valores atipicos, si son errores, o si son datos validos. 

Lo mas relevante al analizar estos valores atipicos es lo siguiente:
hay llamadas de mas de 40 horas, esto es demasiado, y no tiene sentido que sea un valor valido, podrían ser errores en el registro (por ejemplo, una llamada que no se cerró correctamente).
Asumiento que estas llamadas son errores, podrían inflar incorrectamente métricas como el tiempo de espera promedio o la duración total de las llamadas.
Considerando esta informacion colocare un limite en la duracion de las llamadas y eliminare todas las que esten por encima de este rango. Colocare este tope en 8 horas que es el maximo en una jornada (se me ocurre).


Llamadas automáticas o masivas:
Hay un pico de 4817 llamadas en un solo registro, estas llamadas podrían ser válidas (por ejemplo, un operador que realiza muchas llamadas en un corto período). sin embargo esto me sueno un poco ilogico, mas bien pienso que estos valores atipicos podrian representar errores tecnicos en el registro de los datos. 
Es importante revisar bien para decidir como tratarlos pues podrían distorsionar métricas como el número total de llamadas o la duración promedio.

```python
# Definir un límite superior razonable para la duración de las llamadas (en segundos)
limite_superior_call_duration = 8 * 60 * 60  # 8 horas en segundos
limite_superior_total_call_duration = 8 * 60 * 60  # 8 horas en segundos

# Filtrar las llamadas que exceden el límite superior
telecom_data_filtrado = telecom_data[
    (telecom_data['call_duration'] <= limite_superior_call_duration) & 
    (telecom_data['total_call_duration'] <= limite_superior_total_call_duration)
]

# Verificar el número de filas después de la limpieza
print("Número de filas después de eliminar llamadas extremadamente largas:", len(telecom_data_filtrado))
```

```python
# Estadísticas descriptivas después de la limpieza
print("Estadísticas de call_duration después de la limpieza:")
print(telecom_data_filtrado['call_duration'].describe())

print("\nEstadísticas de total_call_duration después de la limpieza:")
print(telecom_data_filtrado['total_call_duration'].describe())
```

Hemos eliminado las llamadas con duracion mayor a 8 horas, en total fueron eliminadas 141 llamadas, lo que es un porcentaje minimo de neustros datos, pero que nos ayuda a mantener los valores en un rango realista, como podemos verlo en la estadisticas descriptivas despues de la limpieza.

```python
# Estadísticas descriptivas de calls_count
print("Estadísticas de calls_count:")
print(telecom_data_filtrado['calls_count'].describe())

# Verificar registros con calls_count extremadamente alto
llamadas_masivas = telecom_data_filtrado[telecom_data_filtrado['calls_count'] > 100]
print("Registros con calls_count > 100:")
print(llamadas_masivas)
```

```python
# Filtrar el registro con calls_count = 4817
llamada_masiva = telecom_data_filtrado[telecom_data_filtrado['calls_count'] == 4817]

# Mostrar el registro
print("Registro con calls_count = 4817:")
print(llamada_masiva)

# Verificar si hay otros registros para el mismo operador en fechas cercanas
operador_masivo = llamada_masiva['operator_id'].values[0]
fecha_masiva = llamada_masiva['date'].values[0]

# Convertir la fecha a un objeto datetime sin zona horaria para evitar errores
fecha_masiva = pd.to_datetime(fecha_masiva).tz_localize(None)

# Filtrar registros para el mismo operador en un rango de fechas cercanas
registros_operador_masivo = telecom_data_filtrado[
    (telecom_data_filtrado['operator_id'] == operador_masivo) & 
    (telecom_data_filtrado['date'].dt.tz_localize(None) >= fecha_masiva - pd.Timedelta(days=7)) & 
    (telecom_data_filtrado['date'].dt.tz_localize(None) <= fecha_masiva + pd.Timedelta(days=7))
]

# Mostrar los registros del operador en fechas cercanas
print("Registros del operador en fechas cercanas:")
print(registros_operador_masivo)
```

```python
# Calcular la diferencia absoluta entre calls_count y 4817
telecom_data_filtrado['diff'] = abs(telecom_data_filtrado['calls_count'] - 4817)

# Obtener los 10 valores más cercanos a 4817
valores_cercanos = telecom_data_filtrado.nsmallest(10, columns=['diff'])

# Mostrar los 10 valores más cercanos
print("Los 10 valores más cercanos a 4817 en calls_count:")
print(valores_cercanos[['user_id', 'date', 'operator_id', 'calls_count']])
```

```python
# Establecer un límite superior razonable para calls_count (por ejemplo, 1000)
limite_superior_calls_count = 1000

# Filtrar las llamadas que exceden el límite superior
telecom_data_filtrado = telecom_data_filtrado[telecom_data_filtrado['calls_count'] <= limite_superior_calls_count]

# Verificar el número de filas después de la limpieza
print("Número de filas después de eliminar llamadas masivas:", len(telecom_data_filtrado))
```

Para poder decidir el tratamiento de estos valores atipicos que pueden afectar significativamente nuestro analisis, he realizado una profundizacion en su naturaleza. 
este valor extremo fue realizado por el operador 929428.0 en un solo día (4,817 llamadas el 11 de noviembre de 2019).
El mismo tiene tambien otros registros con un número alto de llamadas (por ejemplo, 1,533 llamadas el 5 de noviembre de 2019).
Muchas de estas llamadas están marcadas como perdidas, lo que sugiere un problema técnico o un comportamiento inusual.
Despues de este pico el siguiente valor muy alto es de 2,168 llamadas, tambien en un solo dia (el 7 de octubre de 2019).
Estas llamadas también están marcadas como perdidas, lo que sugiere que podrían ser llamadas automáticas o no asociadas a un operador específico.
Considerando esta informacion y sin poder contactar a la empresa para poder indagar sobre estos valores. los considerare un error y por lo tanto voy a eliminarlos estableciendo un límite superior razonable para calls_count de 1000 llamadas por dia y eliminare las que excedan este límite.
Esto nos permitirá eliminar errores obvios sin afectar significativamente el análisis.
realizado este procedimiento, elimine 23 filas, el (0.05% del total).

# Agrupación por operador

Vamos a calcular para cada operator_id:

Total de llamadas atendidas.

Total de llamadas perdidas.

Porcentaje de llamadas perdidas.

Duración promedio de las llamadas.

Total de llamadas salientes realizadas.

# Detección de operadores ineficaces

Recordemos que un operador es ineficaz si:

Tiene una alta tasa de llamadas entrantes perdidas.

Tiene un tiempo de espera prolongado (reflejado en total_call_duration - call_duration).

Si realiza llamadas salientes, su volumen es muy bajo (por ejemplo, menos de 50 llamadas en total).

Vamos a crear tres filtros basados en lo anterior, y luego combinarlos.


1. Volver a calcular las diferencias de espera y separar llamadas entrantes

```python
# Copia del dataset solo con operadores válidos
operadores_validos = telecom_data_filtrado[telecom_data_filtrado['operator_id'] != 'UNKNOWN'].copy()

# Calculamos el tiempo de espera (total - conversación)
operadores_validos['tiempo_espera'] = operadores_validos['total_call_duration'] - operadores_validos['call_duration']

# Filtrar solo llamadas entrantes
entrantes = operadores_validos[operadores_validos['direction'] == 'in']
```

2. Agrupar por operador y calcular métricas de ineficiencia

```python
# Agrupamos por operador solo en llamadas entrantes
ineficiencia_por_operador = entrantes.groupby('operator_id').agg(
    total_entrantes=('calls_count', 'sum'),
    llamadas_perdidas=('is_missed_call', 'sum'),
    tiempo_espera_promedio=('tiempo_espera', 'mean')
)

# Calculamos % de llamadas entrantes perdidas
ineficiencia_por_operador['porcentaje_perdidas'] = (
    ineficiencia_por_operador['llamadas_perdidas'] / ineficiencia_por_operador['total_entrantes'] * 100
).round(2)

# Redondeamos tiempo de espera
ineficiencia_por_operador['tiempo_espera_promedio'] = ineficiencia_por_operador['tiempo_espera_promedio'].round(2)
```

3. Traer llamadas salientes desde el dataset principal

```python
# Agrupamos las llamadas salientes por operador
salientes = operadores_validos[operadores_validos['direction'] == 'out']
salientes_por_operador = salientes.groupby('operator_id')['calls_count'].sum().to_frame()
salientes_por_operador.columns = ['total_salientes']
```

4. Unir todo en un solo DataFrame

```python
# Unimos las métricas de ineficiencia y salientes
operadores_analisis = ineficiencia_por_operador.join(salientes_por_operador, how='left')
operadores_analisis['total_salientes'] = operadores_analisis['total_salientes'].fillna(0)
```

5. Aplicar los filtros de ineficacia

```python
# Criterios de ineficiencia
condicion_perdidas_altas = operadores_analisis['porcentaje_perdidas'] > 30
condicion_espera_alta = operadores_analisis['tiempo_espera_promedio'] > 300  # más de 5 min
condicion_salientes_bajo = operadores_analisis['total_salientes'] < 50

# Operadores que cumplen al menos uno de los criterios
operadores_ineficaces = operadores_analisis[
    condicion_perdidas_altas | condicion_espera_alta | condicion_salientes_bajo
]

print("=== OPERADORES INEFICACES DETECTADOS ===")
display(operadores_ineficaces.sort_values(by='porcentaje_perdidas', ascending=False).head(10))
```

Los operadores detectados presentan al menos uno de los siguientes problemas:

Altos porcentajes de llamadas entrantes perdidas (por ejemplo, 948756.0 con 100%).

Tiempos de espera prolongados antes de que el cliente cuelgue (aunque en este ejemplo, la mayoría están por debajo de 1 min).

Muy bajo volumen de llamadas salientes (por ejemplo, 913886.0 y 934098.0 con 0 llamadas salientes).

# Pruebas de hipótesis estadísticas

Vamos a formular y comprobar una o más hipótesis estadísticas con base en los datos analizados. Estas pruebas nos ayudan a confirmar si ciertas diferencias o patrones observados son estadísticamente significativos o si podrían deberse al azar.

# Hipótesis 1:
 ¿Los operadores que pierden más llamadas tienen tiempos de espera promedio más altos?
 
 Hipótesis nula (H₀): No hay diferencia significativa en el tiempo de espera promedio entre operadores con alta y baja tasa de pérdida.

Hipótesis alternativa (H₁): Los operadores con alta tasa de pérdida tienen tiempos de espera significativamente más altos.

```python
from scipy.stats import ttest_ind

# Separar grupos
grupo_alto = operadores_analisis[operadores_analisis['porcentaje_perdidas'] > 30]['tiempo_espera_promedio']
grupo_bajo = operadores_analisis[operadores_analisis['porcentaje_perdidas'] <= 30]['tiempo_espera_promedio']

# Prueba t de Student (asumimos varianzas distintas)
t_stat, p_value = ttest_ind(grupo_alto, grupo_bajo, equal_var=False)

print("=== PRUEBA DE HIPÓTESIS ===")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-valor: {p_value:.4f}")

# Interpretación
if p_value < 0.05:
    print("Resultado: Rechazamos H₀ — hay diferencia significativa en el tiempo de espera.")
else:
    print("Resultado: No se rechaza H₀ — no hay evidencia suficiente de diferencia.")
```

# Conclusion prueba estadistica #1.

Se formuló una hipótesis estadística para evaluar si los operadores con mayor porcentaje de llamadas entrantes perdidas también tienen un tiempo de espera promedio mayor.

Usamos una prueba t de Student para comparar el tiempo de espera promedio entre:

Grupo A: operadores con más del 30% de llamadas perdidas

Grupo B: operadores con 30% o menos de llamadas perdidas

El resultado fue un T-statistic de -8.34 y un p-valor < 0.0001, lo cual es estadísticamente significativo. Esto nos lleva a rechazar la hipótesis nula.

Concluimos entonces que existe una diferencia real y significativa en el tiempo de espera: los operadores que pierden más llamadas también hacen esperar más a los clientes, lo cual refuerza la necesidad de intervención por parte de los supervisores.



# Hipótesis 2:
¿Los operadores con bajo volumen de llamadas salientes tienen menor duración promedio de conversación?

Esto nos puede ayudar a detectar si los operadores que hacen pocas llamadas también aportan poco en términos de interacción con los clientes (como si solo marcaran sin lograr contacto).

H₀: No hay diferencia significativa en la duración promedio de conversación entre operadores con pocas vs muchas llamadas salientes.

H₁: Los operadores con pocas llamadas salientes tienen duraciones significativamente más bajas.

```python
from scipy.stats import ttest_ind

# Solo hacemos join si la columna no existe
if 'promedio_duracion' not in operadores_analisis.columns:
    operadores_analisis = operadores_analisis.join(operadores_stats[['promedio_duracion']], how='left')

# Separar los grupos según total de llamadas salientes
grupo_bajo_saliente = operadores_analisis[operadores_analisis['total_salientes'] < 50]['promedio_duracion']
grupo_alto_saliente = operadores_analisis[operadores_analisis['total_salientes'] >= 50]['promedio_duracion']

# Prueba t de Student
t_stat2, p_value2 = ttest_ind(grupo_bajo_saliente, grupo_alto_saliente, equal_var=False)

# Mostrar resultados
print("=== PRUEBA DE HIPÓTESIS 2 ===")
print(f"T-statistic: {t_stat2:.2f}")
print(f"P-valor: {p_value2:.4f}")

if p_value2 < 0.05:
    print("Resultado: Rechazamos H₀ — hay diferencia significativa en la duración.")
else:
    print("Resultado: No se rechaza H₀ — no hay evidencia suficiente de diferencia.")

```

# Conclusion de prueba estadistica #2
Se formuló una segunda hipótesis estadística para analizar si los operadores que hacen pocas llamadas salientes también tienen una duración promedio de conversación más baja.

Esta relación es importante porque podría indicar operadores con baja actividad real, que no solo realizan pocas llamadas, sino que también tienen interacciones breves o poco efectivas.

Al aplicar una prueba t de Student, se obtuvo un T-statistic de -11.80 y un p-valor < 0.0001, lo que nos lleva a rechazar la hipótesis nula.

Concluimos que sí existe una diferencia significativa en la duración promedio de conversación: los operadores con bajo volumen de llamadas salientes tienden a tener conversaciones más cortas, lo que puede reforzar su clasificación como ineficaces.

```python
# Filtrar operadores válidos
validos = telecom_data_filtrado[telecom_data_filtrado['operator_id'] != 'UNKNOWN']

# Agrupar por operador y sumar llamadas perdidas
top_perdidas = validos[validos['is_missed_call'] == True] \
    .groupby('operator_id')['calls_count'].sum() \
    .sort_values(ascending=False).head(10)

# Gráfico de barras
plt.figure(figsize=(10, 5))
sns.barplot(x=top_perdidas.values, y=top_perdidas.index, palette="Reds_r")
plt.title('Top 10 Operadores con Más Llamadas Perdidas')
plt.xlabel('Cantidad de llamadas perdidas')
plt.ylabel('ID de operador')
plt.tight_layout()
plt.show()
```

# Interpretacion de la grafica 

Este gráfico muestra los 10 operadores con mayor cantidad de llamadas perdidas, sumadas a lo largo del periodo de análisis.

Algunos operadores presentan más de 6,000 llamadas perdidas, lo que indica graves problemas de atención o capacidad operativa.
En contextos como CallMeMaybe, donde se manejan grandes volúmenes de llamadas entrantes, estas pérdidas representan una gran cantidad de clientes no atendidos, lo cual puede traducirse en pérdidas económicas o mala experiencia del usuario.

Esta visual es clave para detectar a los operadores críticos y reforzar el caso de intervención inmediata o ajuste de estrategia operativa.

```python
# Asegurarse de tener el tiempo de espera calculado
operadores_validos = telecom_data_filtrado[telecom_data_filtrado['operator_id'] != 'UNKNOWN'].copy()
operadores_validos['tiempo_espera'] = operadores_validos['total_call_duration'] - operadores_validos['call_duration']

# Agrupamos por operador para obtener % de llamadas perdidas
ineficiencia = operadores_validos[operadores_validos['direction'] == 'in'].groupby('operator_id').agg(
    total=('calls_count', 'sum'),
    perdidas=('is_missed_call', 'sum')
)
ineficiencia['porcentaje_perdidas'] = ineficiencia['perdidas'] / ineficiencia['total'] * 100

# Asignamos grupo al DataFrame original
grupo_porcentaje = ineficiencia['porcentaje_perdidas']
operadores_validos['porcentaje_perdidas'] = operadores_validos['operator_id'].map(grupo_porcentaje)
operadores_validos = operadores_validos.dropna(subset=['porcentaje_perdidas'])

# Clasificamos operadores
operadores_validos['grupo'] = operadores_validos['porcentaje_perdidas'].apply(
    lambda x: '> 30%' if x > 30 else '≤ 30%'
)

# Boxplot
plt.figure(figsize=(6, 5))
sns.boxplot(data=operadores_validos, x='grupo', y='tiempo_espera', palette='coolwarm')
plt.title('Tiempo de Espera por Grupo de Porcentaje de Pérdidas')
plt.xlabel('Grupo de operador según % de llamadas perdidas')
plt.ylabel('Tiempo de espera (segundos)')
plt.ylim(0, 1500)  # Limitar para mejor visibilidad
plt.tight_layout()
plt.show()
```

# Interpretacion de la grafica 

Esta gráfica compara el tiempo de espera promedio de las llamadas en dos grupos de operadores:
Grupo 1 (≤ 30%): operadores con baja proporción de llamadas perdidas.
Grupo 2 (> 30%): operadores con alta proporción de llamadas perdidas.
A simple vista, se observa que los operadores con mayor porcentaje de pérdidas presentan tiempos de espera considerablemente más bajos. Esto se debe a que sus llamadas no son atendidas, sino que los clientes cuelgan antes.
En contraste, los operadores con menor porcentaje de pérdidas acumulan más tiempo de espera, lo cual indica que sus llamadas tienen más intentos de conexión y atención, incluso si el cliente termina abandonando.
Esta diferencia fue confirmada mediante una prueba estadística (p < 0.0001), lo que refuerza la validez del patrón observado.
